```python
import random
import gym
import numpy as np
from collections import deque
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from colabgymrender.recorder import Recorder
```

State : numpy array (4,)
    High : [4.8 inf 0.42 inf]
    Low : [-4.8 -inf -0.42 -inf]
Action : 	
    0-> Push cart to the left
    1-> Push cart to the right
Reward : +1 even for Termination step as long as Pole is upword

Episod Termination:
 1- If Pole Angle is greater than ±12°
 2- Cart Position is greater than ±2.4 (center of the cart reaches the edge of the display)
 3- Episode length is greater than 500 (200 for v0)

```python
class DQN_Agent:

    # Initializes attributes and constructs Neural Network model 
   
    def __init__(self, state_size, action_size, gamma, epsilon_decay, weight_backup = None):
        self.weight_backup   = weight_backup
        self.state_size      = state_size
        self.action_size     = action_size
        self.memory          = deque(maxlen=2000)
        
        # Hyperparameters
        self.gamma = gamma                  # Discount rate
        self.epsilon = 1.0                  # Exploration rate
        self.epsilon_min = 0.01              # Minimal exploration rate (epsilon-greedy)
        self.epsilon_decay = epsilon_decay  # Decay rate for epsilon
        self.update_rate = 1000             # Number of steps until updating the target network
        
        # Construct DQN models
        self.learning_rate = 0.001
        self.model = self._build_model() 
        # self.target_model = self._build_model()
        # self.target_model.set_weights(self.model.get_weights())
        # self.model.summary()

    
    # Build Neural Network Model
    def _build_model(self):
        model = Sequential()
        model.add(Dense(24, input_dim=self.state_size, activation='relu'))
        model.add(Dense(24, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse',
                      optimizer=Adam(lr=self.learning_rate))
        if self.weight_backup != None and os.path.isfile(self.weight_backup):
            model.load_weights(self.weight_backup)
            self.epsilon = self.epsilon_min
        return model

  
    # Stores experience in replay memory
    def store_experience(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    
    # Agent Choose action based on epsilon-greedy policy
    def choose_action(self, state):
        # Random Exploration
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        # Expotation
        act_values = self.model.predict(state)
        return np.argmax(act_values[0]) 

  
    # Trains the model using randomly selected experiences in the replay memory
    def experience_replay(self, batch_size):

        if len(self.memory) < batch_size:
            return
        # sample random mini-batches of transition
        sample_batch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in sample_batch:
            target = reward
            if not done:
              target = reward + self.gamma * np.amax(self.model.predict(next_state)[0])
            # Construct the target vector as follows:
            # Use the current model to output the Q-value predictions
            target_f = self.model.predict(state)
            # Rewrite the chosen action value with the computed target
            target_f[0][action] = target
             # Use vectors to fit model
            self.model.fit(state, target_f, epochs=1, verbose=0)
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay


    # Sets the target model parameters to the current model parameters
    def update_target_model(self):
        self.target_model.set_weights(self.model.get_weights())
            
  
    # Loads a saved model
    def load(self, name):
        self.model.load_weights(name)

    # Saves parameters of a trained model
    def save(self, name):
        self.model.save(name)
```

```python
import os
env = gym.make('CartPole-v1')
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
agent = DQN_Agent(state_size, action_size, gamma = 0.95, epsilon_decay =0.995, weight_backup = "cartpole_weight.h5")
done = False
batch_size = 32
```

```python
for e in range(500):
    state = env.reset()
    state = np.reshape(state, [1, agent.state_size])
    
    done = False
    index = 0
    total_reward = 0
    while not done:
        action = agent.choose_action(state)

        next_state, reward, done, _ = env.step(action)
        next_state = np.reshape(next_state, [1, agent.state_size])
        agent.store_experience(state, action, reward, next_state, done)
        state = next_state
        # total_reward += reward 
        index += 1
    print("Episode {}# Score: {}".format(e, index + 1))
    agent.experience_replay(batch_size)
```

```python
def evaluate_agent_episodes(env, n_eval_episodes):
  episode_rewards = []

  directory = "./video"
  env = Recorder(env, directory)  
  for episode in range(n_eval_episodes):
    state = env.reset()
    done = False
    total_rewards_ep = 0
    
    while not done:
      # Take the action (index) that have the maximum expected future reward given that state
      # action =  agent.choose_action(state)
      state =  np.reshape(next_state, [1, state_size])
      action = np.argmax(agent.model.predict(state)[0])
      new_state, reward, done, info = env.step(action)
      total_rewards_ep += reward
      
        
      if done:
        break
      state = new_state

    episode_rewards.append(total_rewards_ep)
  env.play()
  return episode_rewards
```
